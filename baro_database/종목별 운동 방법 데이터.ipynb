#### 3. 강도

#### 운동강도 측정법 
저강도(Light-intensity) <br>
1.5 ~ 2.9 METs 강도(휴식 시의 1.5 ~ 2.9배 강도)를 말하며, 운동 중 호흡의 변화를 거의 느끼지 못하고 대화나 노래 부르기가 가능한 편안한 수준으로 정의됩니다.<br>

중강도(moderate-intensity) <br>
3~6METs 강도(휴식시의 3~6배 강도)를 말하며 운동 중 숨이 약간 가쁘지만 대화가 가능할 정도의 수준으로 정의된다. <br>

고강도(vigorous-intensity)<br>
6METs 이상의 강도(휴식시의 6배 강도)를 말하며 운동 중 대화가 불가능 할 정도의 숨이 매우 가쁜 정도의 수준으로 정의된다.<br>
